2.What are the types of random variables?

###  **Types of Random Variables**

A **random variable (RV)** is a variable whose value is determined by the outcome of a random experiment. Random variables are classified based on the **type of values** they can take.


###  **1. Discrete Random Variable**

* **Definition:** Takes **countable, distinct values** (often integers).
* **Values:** Finite or countably infinite.
* **Probability Representation:** Probability Mass Function (PMF)
* **Examples:**

  * Number of heads in 5 coin tosses
  * Number of defective items in a batch
  * Number of customers visiting a store in a day


###  **2. Continuous Random Variable**

* **Definition:** Takes **any value within a continuous range**.
* **Values:** Infinite possibilities within an interval.
* **Probability Representation:** Probability Density Function (PDF)
* **Examples:**

  * Height of a person
  * Time taken to complete a task
  * Weight of an object


###  **3. Mixed (or Hybrid) Random Variable**

* **Definition:** A variable that has **both discrete and continuous components**.
* **Examples:**

  * Insurance claims where some people have **zero claims** (discrete) and others have **any positive amount** (continuous)
  * Time until a machine fails where some fail immediately and others over a continuous period

###  **Key Differences Between Discrete and Continuous RVs**

| Feature                  | Discrete RV              | Continuous RV        |
| ------------------------ | ------------------------ | -------------------- |
| **Values**               | Countable                | Any value in a range |
| **Probability Function** | PMF                      | PDF                  |
| **Sum/Integral**         | Sum of probabilities = 1 | Integral of PDF = 1  |
| **Example**              | Dice roll, # of defects  | Height, weight, time |

* **Example:**

###  **Key Differences**

| Feature                  | Discrete Distribution                   | Continuous Distribution                     |

6.What is the Central Limit Theorem (CLT), and why is it critical in statistics?

###  **Central Limit Theorem (CLT) ‚Äî Definition & Importance**


###  **Definition**

The **Central Limit Theorem (CLT)** states that:

> When independent random samples of a sufficiently large size are drawn from any population (regardless of its original distribution), the **sampling distribution of the sample mean** will tend to follow a **normal distribution**, with:
>
> * Mean = population mean (Œº)
> * Standard deviation = ( \frac{\sigma}{\sqrt{n}} ), where ( n ) = sample size


###  **Example**

Suppose we record the daily number of customers at a store (which may not be normally distributed).
If we take multiple samples of 50 days and calculate the average number of customers in each sample, the **distribution of those averages** will be **approximately normal**.


###  **Why CLT is Critical in Statistics**

| **Reason**                                  | **Explanation**                                                                                                                |
| ------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------ |
| **1. Foundation of Inferential Statistics** | It allows us to use **normal probability theory** to make inferences about population parameters (like mean, proportion).      |
| **2. Enables Confidence Intervals**         | Since sampling means are normally distributed, we can estimate **confidence intervals** for the population mean.               |
| **3. Basis for Hypothesis Testing**         | Many tests (z-test, t-test, ANOVA) rely on the normality assumption, justified by CLT.                                         |
| **4. Works for Non-Normal Data**            | Even if the original population is **not normal**, CLT ensures that the sample means are approximately normal for large ( n ). |
| **5. Simplifies Real-World Problems**       | It lets analysts model real data with normal distribution formulas, even if raw data is complex or skewed.                     |


###  **Visualization Idea**

If you simulate samples from any skewed distribution (e.g., exponential), the histogram of **sample means** will still look **bell-shaped** ‚Äî that‚Äôs CLT in action.




7.What is the significance of confidence intervals in statistical analysis?

###  **Significance of Confidence Intervals in Statistical Analysis**

A **Confidence Interval (CI)** provides a **range of values** that is likely to contain the **true population parameter** (like mean, proportion, or variance) with a certain **level of confidence** (commonly 95% or 99%).

It gives us **both an estimate and the uncertainty** associated with that estimate.


###  **1. Meaning of Confidence Interval**

A confidence interval does **not** say that the true value definitely lies inside the interval ‚Äî instead, it means:

> ‚ÄúIf we were to take many random samples and compute a confidence interval each time, about 95% of those intervals would contain the true population parameter.‚Äù



###  **3. Example**

Suppose the average daily sales in a store are ‚Çπ250 with a standard deviation of ‚Çπ20 based on a sample of 25 days.


 95% Confidence Interval = **(242.16, 257.84)**

Interpretation ‚Üí We are 95% confident that the **true average daily sales** lie between ‚Çπ242.16 and ‚Çπ257.84.

---

###  **4. Significance / Importance**

| **Purpose**                     | **Explanation**                                                                              |
| ------------------------------- | -------------------------------------------------------------------------------------------- |
| **Estimates uncertainty**       | CI gives a range rather than a single number, showing how precise your estimate is.          |
| **Reliability indicator**       | A **narrow CI** indicates high precision; a **wide CI** shows more uncertainty.              |
| **Decision-making tool**        | Helps managers or researchers decide if a result is statistically meaningful.                |
| **Comparison aid**              | Overlapping CIs between two groups suggest no significant difference.                        |
| **Supports hypothesis testing** | If a hypothesized value (e.g., mean = 0) lies outside the CI, we can reject that hypothesis. |



8.What is the concept of expected value in a probability distribution?

###  **Concept of Expected Value in a Probability Distribution**

The **Expected Value (EV)** ‚Äî also known as the **mean** of a probability distribution ‚Äî represents the **long-term average outcome** of a random variable after many repetitions of an experiment.


###  **Definition**

For a random variable ( X ), the **expected value** is the **weighted average** of all possible values that ( X ) can take, where each value is weighted by its probability.



###  **Intuitive Meaning**

The expected value is what you **expect on average** if you repeat the random experiment a large number of times.


###  **Example**

| X (Outcome) | P(X) |
| ----------- | ---- |
| 1           | 0.2  |
| 2           | 0.5  |
| 3           | 0.3  |

[
E(X) = (1)(0.2) + (2)(0.5) + (3)(0.3) = 0.2 + 1.0 + 0.9 = 2.1
]

 The **expected value = 2.1**, meaning on average, the random variable ( X ) will take the value **2.1** over many trials.



###  **Real-Life Examples**

* In gambling: the expected value of a bet tells you **the average gain or loss** per play.
* In business: expected value helps in **risk analysis** and **decision-making** (e.g., expected profit).
* In statistics: it represents the **mean** of a probability distribution.



````python
'''9 Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.
'''
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
n = 1000

data = np.random.normal(mean, std_dev, n)

# Step 2: Compute mean and standard deviation using NumPy
calculated_mean = np.mean(data)
calculated_std = np.std(data, ddof=1)  # sample standard deviation

# Step 3: Display the results
print(f"Calculated Mean: {calculated_mean:.2f}")
print(f"Calculated Standard Deviation: {calculated_std:.2f}")

# Step 4: Draw a histogram to visualize the distribution
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Normally Distributed Data (Œº=50, œÉ=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(calculated_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {calculated_mean:.2f}")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

````

###  **Concept: Applying the Central Limit Theorem (CLT)**

The **Central Limit Theorem (CLT)** states that:

> When we take a large number of random samples from a population, the sampling distribution of the sample means will be approximately **normal**, regardless of the population‚Äôs distribution ‚Äî **as long as the sample size is large enough (n ‚â• 30)**.

Even if our dataset is small (like here, n = 20), we can **assume approximate normality** and use the **t-distribution** instead of the z-distribution to build a confidence interval.


###  **Steps to Estimate Average Sales with 95% Confidence Interval**

1. **Compute sample mean (xÃÑ)** ‚Äî the average of all sales.
2. **Compute sample standard deviation (s)**.
3. **Find the t-critical value** for 95% confidence and `df = n - 1`.
4. **Compute Standard Error (SE)** = s / ‚àön
5. **Calculate Confidence Interval (CI):**

   [
   CI = xÃÑ ¬± t_{critical} \times SE
   ]


###  **Python Code Implementation**

```python
import numpy as np
import scipy.stats as stats

# Given data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Step 1: Calculate sample statistics
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # sample standard deviation
n = len(daily_sales)

# Step 2: Standard Error
standard_error = std_sales / np.sqrt(n)

# Step 3: t-critical value for 95% confidence interval
confidence_level = 0.95
df = n - 1
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

# Step 4: Margin of Error
margin_of_error = t_critical * standard_error

# Step 5: Confidence Interval
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
