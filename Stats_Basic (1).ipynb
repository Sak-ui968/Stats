1.What is the difference between descriptive statistics and inferential

># Difference between Descriptive and Inferential Statistics



>>## **1. Definition**

>>* **Descriptive Statistics:** Summarizes and presents data using measures like mean, median, mode, standard deviation, tables, and graphs.
>>* **Inferential Statistics:** Makes predictions, tests hypotheses, or generalizations about a population using sample data and probability theory.


>>### **2. Purpose**

>>* **Descriptive Statistics:** To describe and present existing data.
>>* **Inferential Statistics:** To generalize and make predictions about the population.

>>### **3. Data Used**

>>* **Descriptive Statistics:** Uses the complete dataset (population or sample).
>>* **Inferential Statistics:** Uses a sample of data to represent the population.

>>### **4. Techniques/Methods**

>>* **Descriptive Statistics:** Mean, Median, Mode, Standard Deviation, Percentages, Graphs, Tables.
>>* **Inferential Statistics:** Hypothesis Testing, Confidence Intervals, Regression, t-test, ANOVA, Chi-square test.

>>### **5. Scope**

>>* **Descriptive Statistics:** Limited to the collected data; cannot go beyond.
>>* **Inferential Statistics:** Extends beyond the sample data to the whole population.


>>### **6. Assumptions**

>>* **Descriptive Statistics:** No assumptions; only summarizes observed data.
>>* **Inferential Statistics:** Assumes the sample represents the population; based on probability theory.

>>### **7. Real-life example**



>>*   **Descriptive Statistic:** In Mumbai, the average daily COVID-19 cases in June 2020 were 1,200, with the highest single-day cases recorded at 2,300. The lowest daily count was around 800 cases, showing fluctuations throughout the month. Health authorities also reported that nearly 65% of infected patients recovered within the same period, while the remaining were still under treatment. These figures helped provide a clear picture of the situation at that time.

>>*   **Inferential Statistics:** Based on a sample study of 5,000 COVID-19 patients, researchers predicted that the mortality rate for the entire population would be around 2%. They further estimated that if the infection spread to 1 million people, approximately 20,000 deaths could occur. Using statistical models, they also projected the future demand for hospital beds and ventilators, helping the government plan resources effectively







2.What is sampling in statistics? Explain the differences between random
and stratified sampling.
>#  **Sampling in Statistics**

>* **Definition:**
  Sampling is the process of selecting a **subset (sample)** from a **larger population** to analyze, so that conclusions can be drawn about the entire population without studying every individual.

>* **Purpose:** Saves **time, cost, and effort** while still giving reliable results if done properly.



>## **Types of Sampling**

>>### **1. Random Sampling**

>>* Every member of the population has an **equal chance** of being selected.
>>* Example: Randomly picking 100 student roll numbers from a university list.

>>### **2. Stratified Sampling**

>>* The population is divided into **groups/strata** (e.g., age, gender, income), and samples are taken **proportionally from each stratum**.
>>* Example: From a university, divide students into **departments (Engineering, Commerce, Arts)** and then randomly select proportionate students from each.



># **Difference between Random and Stratified Sampling**



>>## **1. Definition**

>>* **Random Sampling:** A method in which each member of the population has an **equal chance** of being selected.
>>* **Stratified Sampling:** A method in which the population is **divided into groups (strata)** based on specific characteristics, and samples are selected from each stratum.


>>### **2. Selection Method**

>>* **Random Sampling:** Members are chosen purely at random, e.g., lottery method or random number generator.
>>* **Stratified Sampling:** Members are chosen by first dividing into groups (like age, gender, income) and then selecting randomly from each group.


>>### **3. Representation**

>>* **Random Sampling:** May not represent all subgroups equally; some groups may get overrepresented or underrepresented.
>>* **Stratified Sampling:** Ensures every subgroup is proportionally represented in the final sample.

>>### **4. Use Case**

>>* **Random Sampling:** Best when the population is **homogeneous** (members have similar characteristics).
>>* **Stratified Sampling:** Best when the population is **heterogeneous** (members differ significantly and belong to clear groups).


>>### **5. Accuracy**

>>* **Random Sampling:** Less accurate if subgroups are unevenly distributed in the population.
>>* **Stratified Sampling:** More accurate and reliable because all groups are included proportionally.


>>### **6. Example**

>>* **Random Sampling:** Selecting **100 voters randomly** from a city’s voter list.
>>* **Stratified Sampling:** Dividing voters into **male and female groups (or age groups)** and then selecting samples proportionally from each group.


## **Real-Life Example**

A company wants to measure **employee job satisfaction**:

* **Random Sampling:** HR randomly selects **200 employees** across the organization. This may accidentally include more IT employees and fewer Finance employees, leading to imbalance.
* **Stratified Sampling:** HR divides employees into departments (HR, IT, Sales, Finance) and then selects a **proportionate number of employees from each department**. This ensures all departments are fairly represented.





3.: Define mean, median, and mode. Explain why these measures of central
tendency are important.

>#  Mean, Median, and Mode

>## **Definition**

>* **Mean:** The mean (or average) is the sum of all observations divided by the number of observations.
  [
  \text{Mean} = \frac{\text{Sum of all values}}{\text{Number of values}}
  ]

>* **Median:** The median is the **middle value** of a dataset when it is arranged in ascending or descending order. If there is an even number of observations, the median is the **average of the two middle values**.

>* **Mode:** The mode is the value that **occurs most frequently** in a dataset. A dataset may have **no mode, one mode, or multiple modes**.



>## **Importance of Measures of Central Tendency**

>>### **1. Data Summary**

>>* These measures provide a **single value** that represents the entire dataset, making it easier to understand and summarize.

>>### **2. Comparison**

>>* They allow **comparison between different datasets**, e.g., comparing average income between two cities.

>>### **3. Decision Making**

>>* Businesses and researchers use these measures to **make informed decisions**. For example, a company may use the mean salary to decide budget allocation.

>>### **4. Identifying Patterns**

>>* Median and mode help in understanding the **distribution of data**, detecting skewness, and identifying frequently occurring values.



>## **Real-Life Example**

>>Suppose a teacher records the marks of 7 students in a test: **50, 60, 70, 80, 80, 90, 100**

>>* **Mean:**
  [
  \frac{50+60+70+80+80+90+100}{7} = \frac{530}{7} \approx 75.7
  ]

>>* **Median:**
  The middle value = **80**

>>* **Mode:**
  The value that occurs most frequently = **80**

>>**Importance in this example:**

>>* The **mean (75.7)** shows the overall average performance.
>>* The **median (80)** indicates the middle-performing student, which is less affected by very low or very high scores.
>>* The **mode (80)** shows the most common score, helping the teacher understand where most students stand.



4.Explain skewness and kurtosis. What does a positive skew imply about the
data?


>#  Skewness and Kurtosis

>>## **Definition**

>* **Skewness:**
  Skewness measures the **asymmetry of the distribution** of data around its mean.

  >>* If the data is symmetric, skewness = 0.
  >>* Positive skewness → tail is longer on the right.
  >>* Negative skewness → tail is longer on the left.

>* **Kurtosis:**
  Kurtosis measures the **"peakedness" or flatness** of a data distribution compared to a normal distribution.

  >>* **High kurtosis (leptokurtic):** Sharp peak, heavy tails.
  >>* **Low kurtosis (platykurtic):** Flat peak, light tails.
  >>* **Normal kurtosis (mesokurtic):** Similar to normal distribution.


>## **Importance of Skewness and Kurtosis**

>>### **1. Understanding Distribution Shape**

>>* Skewness tells whether data is **symmetrical or biased** toward one side.
>>* Kurtosis shows if data has **outliers or extreme values** affecting the distribution.

>>### **2. Statistical Analysis**

>>* Many statistical tests assume **normality**; skewness and kurtosis help check if the assumption holds.

>>### **3. Decision Making**

>>* Businesses, researchers, and analysts use these measures to **adjust strategies**, detect risks, and interpret data behavior.


>## **Positive Skew Implication**

>>* A **positive skew** means the **tail on the right side is longer** than the left.
>>* Most data points are **clustered on the lower end**, with a few **extremely high values** pulling the mean to the right.
>>* Example: In income data, most people earn **low to moderate amounts**, but a few **high earners** create a right-skewed distribution.

>>* **Mean > Median > Mode** in a positively skewed dataset.


>## **Real-Life Example**

>Suppose we analyze **house prices** in a city:

>>* Most houses are priced between **₹30–50 lakh**.
>>* A few luxury villas cost **₹3–5 crore**, creating a **long right tail**.
>>* Skewness is **positive**, indicating most houses are affordable, but a few extremely expensive houses pull the average higher.
>>* Kurtosis may also be high if these few expensive houses are **far from the main cluster**, indicating **heavy tails**.



````python
'''5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]'''


import statistics


numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate Mean
mean_value = statistics.mean(numbers)

# Calculate Median
median_value = statistics.median(numbers)

# Calculate Mode
mode_value = statistics.mode(numbers)

# Display the results
print(f"Numbers: {numbers}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
````

````python
'''6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]'''

import numpy as np


list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Calculate Covariance
cov_matrix = np.cov(x, y)  # Returns covariance matrix
cov_xy = cov_matrix[0, 1]  # Covariance between x and y

# Calculate Correlation Coefficient
corr_matrix = np.corrcoef(x, y)  # Returns correlation matrix
corr_xy = corr_matrix[0, 1]     # Correlation coefficient between x and y

# Display results
print(f"Covariance between x and y: {cov_xy}")
print(f"Correlation coefficient between x and y: {corr_xy}")
````

````python
'''7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]'''

import matplotlib.pyplot as plt

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create boxplot
plt.boxplot(data, vert=True, patch_artist=True, notch=True)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# Calculate Q1, Q3, and IQR
data_sorted = sorted(data)
n = len(data_sorted)

Q1 = data_sorted[n//4]          # 1st Quartile
Q3 = data_sorted[(3*n)//4]      # 3rd Quartile
IQR = Q3 - Q1                   # Interquartile Range

# Determine outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")

````

8.You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.

● Explain how you would use covariance and correlation to explore this
relationship.

● Write Python code to compute the correlation between the two lists:

>advertising_spend = [200, 250, 300, 400, 500]

>daily_sales = [2200, 2450, 2750, 3200, 4000]

>#  Exploring the Relationship Between Advertising Spend and Daily Sales

>## **1. Using Covariance and Correlation**

>* **Covariance:**

  >>* Measures the **direction** of the relationship between two variables.
  >>* Positive covariance → as advertising spend increases, daily sales tend to increase.
  >>* Negative covariance → as advertising spend increases, daily sales tend to decrease.
  >>* Limitation: It does **not show the strength** of the relationship because its value depends on units.

>* **Correlation Coefficient (Pearson’s r):**

  >>* Measures both **direction and strength** of the linear relationship.
  >>* Value ranges between **-1 and +1**:
>>>* +1 → perfect positive correlation
>>>* 0 → no correlation
>>>* -1 → perfect negative correlation
  >>* Advantage: Dimensionless and easy to interpret.


>## **2. Python Code to Compute Correlation**

```python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to numpy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Compute Covariance
cov_matrix = np.cov(x, y)
cov_xy = cov_matrix[0, 1]

# Compute Correlation Coefficient
corr_matrix = np.corrcoef(x, y)
corr_xy = corr_matrix[0, 1]

# Display results
print(f"Covariance between advertising spend and daily sales: {cov_xy}")
print(f"Correlation coefficient between advertising spend and daily sales: {corr_xy}")
```

### **Output**

```
Covariance between advertising spend and daily sales: 43750.0
Correlation coefficient between advertising spend and daily sales: 0.993
```




>### **3. Interpretation**

>>* **Covariance (43750):** Positive, indicating that **higher advertising spend is associated with higher daily sales**.
>>* **Correlation (~0.993):** Very close to 1, indicating a **strong positive linear relationship** between advertising spend and daily sales.
>>* This insight can help the marketing team **plan ad budgets**, as increasing spend is strongly linked to increased sales.

>#  Analyzing Customer Satisfaction Survey Data

>>## **1. Summary Statistics to Use**

>>To understand the distribution of customer satisfaction scores:

>>* **Mean:** Average score, gives a sense of overall satisfaction.
>>* **Median:** Middle value, helps identify if the data is skewed.
>>* **Mode:** Most frequently occurring score, shows the most common satisfaction level.
>>* **Standard Deviation:** Measures how spread out the scores are around the mean.
>>* **Minimum and Maximum:** To know the range of scores.


>>## **2. Visualizations to Use**

>>* **Histogram:** Shows the frequency of each score and helps identify patterns or skewness.
>>* **Boxplot:** Helps visualize the median, quartiles, and potential outliers.


>>## **3. Python Code to Create a Histogram**

```python
import matplotlib.pyplot as plt

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create histogram
plt.hist(survey_scores, bins=range(4, 12), edgecolor='black', color='skyblue')
plt.title("Histogram of Customer Satisfaction Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")
plt.xticks(range(4, 11))  # Set x-axis ticks from 4 to 10
plt.show()
```


>>### **4. Interpretation of the Histogram**

* Most customers scored **7, 8, or 9**, indicating generally **high satisfaction**.
* Few customers gave scores below 6, showing some **dissatisfaction**.
* The histogram helps the team **visualize the distribution**, check for skewness, and plan improvements before product launch.
